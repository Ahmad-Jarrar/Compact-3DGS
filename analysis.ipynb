```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd = pd.read_csv('all_results.csv')
```

```python
# filter Runs with names containing bicycle or bonsai

pd_bicycle = pd[pd['Run'].str.contains('bicycle')]
pd_bonsai = pd[pd['Run'].str.contains('bonsai')]
```

```python
pd_bicycle
```

```python
pd_bonsai
```

```python
# plot bar chart Run against SSIM, PSNR, LPIPS

# SSIM
plt.figure(figsize=(15, 5))
plt.bar(pd_bicycle['Run'], pd_bicycle['SSIM'])
# Tilt the x-axis labels
plt.xticks(rotation=90)
plt.xlabel('Run')
plt.ylabel('SSIM')
plt.title('SSIM')

# PSNR
plt.figure(figsize=(15, 5))
plt.bar(pd_bicycle['Run'], pd_bicycle['PSNR'])
# Tilt the x-axis labels
plt.xticks(rotation=90)
plt.xlabel('Run')
plt.ylabel('PSNR')
plt.title('PSNR')

# LPIPS
plt.figure(figsize=(15, 5))
plt.bar(pd_bicycle['Run'], pd_bicycle['LPIPS'])
# Tilt the x-axis labels
plt.xticks(rotation=90)
plt.xlabel('Run')
plt.ylabel('LPIPS')
plt.title('LPIPS')

plt.show()
# plt.figure(figsize=(15, 5))
# plt.bar(pd_bicycle['Run'], pd_bicycle['SSIM'])
# # Tilt the x-axis labels
# plt.xticks(rotation=90)
# plt.xlabel('Run')
# plt.ylabel('SSIM')

```
